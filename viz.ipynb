```python
import altair as alt
import pandas as pd
```

```python
# seconds
entries = {
    "Python Native UDF": 26.701,  # seconds
    "Python Native UDF - Numba JIT": 23.443,  # seconds
    "Vectorized UDF Numba JIT": 2.998,  # seconds
    "Vectorized UDF Rust-based": 2.566,  # seconds
    "Pure SQL": 0.3479,  # seconds
    "Export to Numpy - OpenMP": 1.715,  # seconds
    "Export to Numpy - CUDA": 1.803,  # seconds
}

def to_df(entries):
    data = [[k,v] for k,v in entries.items()]
    cols = ["approach", "time_taken"]
    df = pd.DataFrame(data, columns=cols)
    return df

def get_chart(entries):
    df = to_df(entries)
    min_val = 0
    max_val = df["time_taken"].max()
    return alt.Chart(df).mark_bar().encode(
        alt.Y("approach:N",axis=alt.Axis(title=None) ).sort("-x"),
        alt.X("time_taken:Q", axis=alt.Axis(title='Time Taken')),
        alt.Color("time_taken", scale=alt.Scale(domain=(min_val, 30), scheme="lighttealblue"), legend=None),
        alt.OpacityValue(0.8)
    ).properties(
        width=800,
        height=100
    ).configure_scale(
        bandPaddingInner=0.01
    ).configure(
        autosize=alt.AutoSizeParams(resize=True)
    )
```

```python
entries = {
    "Python Native UDF": 26.701,  # seconds
    "Python Native UDF - Numba JIT": 23.443,  # seconds
}
get_chart(entries)
```
